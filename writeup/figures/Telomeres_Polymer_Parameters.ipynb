# Finding the polymeric properties of telomeres

In this notebook I use the log-likelihood function that's obtained from the polymer simulations and measured telomere datasets to plot the parameter space that best agrees with the measured size data.

## Setup the environment

```python
# The usual suspects
import numpy as np
import matplotlib.pyplot as plt

from rw_helpers import unpackLLH, loadModel

# Plot matplotlib figures in this document.
%matplotlib inline

# Dictionary of results
results = {'Hela L' : {'LLH' : None,
                       'packRatio' : None,
                       'pLength' : None},
           'Hela S' : {'LLH' : None,
                       'packRatio' : None,
                       'pLength' : None}}
```

## Read in the log-likelihood data

```python
# Read in Hela L
currDataset = 'llh_Original_Data_L_dataset_RgTrans.npy'

with open(currDataset, mode = 'rb') as inFile:
    hl = np.load(inFile)

# Unpack the dataset
results['Hela L']['packRatio']   = hl['f0']
results['Hela L']['pLength']     = hl['f1']
(_, _, results['Hela L']['LLH']) = unpackLLH(hl)
```

```python
# Read in Hela S
currDataset = 'llh_Original_Data_S_dataset_RgTrans.npy'

with open(currDataset, mode = 'rb') as inFile:
    hs = np.load(inFile)

# Unpack the dataset
results['Hela S']['packRatio']   = hs['f0']
results['Hela S']['pLength']     = hs['f1']
(_, _, results['Hela S']['LLH']) = unpackLLH(hs)
```

## Plot the log-likelihood

### Hela L

```python
options = {'vmin'   : -1.6e4,
           'vmax'   : results['Hela L']['LLH'].max(),
           'origin' : 'lower',
           'extent' : [results['Hela L']['packRatio'].min(), results['Hela L']['packRatio'].max(),
                       results['Hela L']['pLength'].min(),   results['Hela L']['pLength'].max()],
           'aspect' : 'auto'}

fig = plt.figure(figsize = (3.46, 2.5), dpi = 300)

# Log-liklihood map
plt.imshow(results['Hela L']['LLH'], **options)
plt.colorbar(label = 'Log-likelihood')

# Plot simulated points
plt.scatter(results['Hela L']['packRatio'], 
            results['Hela L']['pLength'],
            s = 2,
            c = 'black')
plt.xlim((results['Hela L']['packRatio'].min(), results['Hela L']['packRatio'].max()))
plt.ylim((results['Hela L']['pLength'].min(),   results['Hela L']['pLength'].max()))
plt.xlabel('Packing ratio, bp/nm')
plt.ylabel('Persistance length, nm')
plt.savefig('HelaL_params.pdf')
```

### Hela S

```python
options = {'vmin'   : -1.6e4,
           'vmax'   : results['Hela S']['LLH'].max(),
           'origin' : 'lower',
           'extent' : [results['Hela S']['packRatio'].min(), results['Hela S']['packRatio'].max(),
                       results['Hela S']['pLength'].min(),   results['Hela S']['pLength'].max()],
           'aspect' : 'auto'}

fig = plt.figure(figsize = (3.46, 2.5), dpi = 300)

# Log-liklihood map
plt.imshow(results['Hela S']['LLH'], **options)
plt.colorbar(label = 'Log-likelihood')

# Plot simulated points
plt.scatter(results['Hela S']['packRatio'], 
            results['Hela S']['pLength'],
            s = 2,
            c = 'black')
plt.xlim((results['Hela S']['packRatio'].min(), results['Hela S']['packRatio'].max()))
plt.ylim((results['Hela S']['pLength'].min(),   results['Hela S']['pLength'].max()))
plt.xlabel('Packing ratio, bp/nm')
plt.ylabel('Persistance length, nm')
plt.savefig('HelaS_params.pdf')
```

## Plot the parameter space close to the mean

The log-likelihood gives us a relative measure of the similarity between the measured distribution and the simulated one. However, higher log-likelihood values may not necessarily correspond to better simulated distributions. We should also consider looking at all simulated distributions with means that are a certain distance from the experimental mean, say +/- 5 nm. This is a very rough estimate of the day-to-day variation in the mean values of the distributions.

```python
import matplotlib.font_manager as font_manager
from matplotlib import rcParams

fontpath = '/usr/share/fonts/truetype/msttcorefonts/Arial.ttf'
prop = font_manager.FontProperties(fname = fontpath)
rcParams['font.family'] = prop.get_name()
rcParams.update({'font.size' : 7})
rcParams.update({'legend.fontsize' : 7})
rcParams.update({'axes.titlesize' : 7})
```

```python
from scipy.stats import sem

# Experimental distributions
LExpDist = np.loadtxt('saved_distrs/Original_Data_L_dataset_RgTrans.txt')
SExpDist = np.loadtxt('saved_distrs/Original_Data_S_dataset_RgTrans.txt')

LMean = np.mean(LExpDist)
SMean = np.mean(SExpDist)

LSEM = sem(LExpDist)
SSEM = sem(SExpDist)

# Model distributions
LModelDists = loadModel(['rw_2015-1-26_HelaL_WT'])
SModelDists = loadModel(['rw_2015-2-2_HelaS_WT'])

LModelMeans = {k: np.mean(v[1]) for k, v in LModelDists.items()}
SModelMeans = {k: np.mean(v[1]) for k, v in SModelDists.items()}
```

```python
# Set the +/- amount around the experimental mean for filtering out incorrect parameter space values
err = 5

# Filter out datasets with means far from the experimental mean
LMeansFiltered = {k: v for k, v in LModelMeans.items() if np.abs(LMean - v) < err}
SMeansFiltered = {k: v for k, v in SModelMeans.items() if np.abs(SMean - v) < err}

# Convert the filtered parameter space values into arrays for plotting
LGoodParams = list(LMeansFiltered.keys())
SGoodParams = list(SMeansFiltered.keys())
```

```python
packRatio, pLength = zip(*LGoodParams)

options = {'vmin'   : -1.6e4,
           'vmax'   : results['Hela L']['LLH'].max(),
           'origin' : 'lower',
           'extent' : [results['Hela L']['packRatio'].min(), results['Hela L']['packRatio'].max(),
                       results['Hela L']['pLength'].min(),   results['Hela L']['pLength'].max()],
           'aspect' : 'auto'}

fig = plt.figure(figsize = (3.46, 2.5), dpi = 300)

# Log-liklihood map
plt.imshow(results['Hela L']['LLH'], **options)
plt.colorbar(label = 'Log-likelihood')

# Plot simulated points
props = {'alpha' : 0.6,
         'edgecolors' : None}
plt.scatter(packRatio, 
            pLength,
            s = 5,
            c = 'white',
            **props)
plt.xlim((results['Hela L']['packRatio'].min(), results['Hela L']['packRatio'].max()))
plt.ylim((results['Hela L']['pLength'].min(),   results['Hela L']['pLength'].max()))
plt.xlabel('Packing ratio, bp/nm')
plt.ylabel('Persistance length, nm')
plt.savefig('HelaL_params_means.pdf')
```

```python
packRatio, pLength = zip(*SGoodParams)

options = {'vmin'   : -1.6e4,
           'vmax'   : results['Hela S']['LLH'].max(),
           'origin' : 'lower',
           'extent' : [results['Hela S']['packRatio'].min(), results['Hela S']['packRatio'].max(),
                       results['Hela S']['pLength'].min(),   results['Hela S']['pLength'].max()],
           'aspect' : 'auto'}

fig = plt.figure(figsize = (3.46, 2.5), dpi = 300)

# Log-liklihood map
plt.imshow(results['Hela S']['LLH'], **options)
plt.colorbar(label = 'Log-likelihood')

# Plot simulated points
props = {'alpha' : 0.6,
         'edgecolors' : None}
plt.scatter(packRatio, 
            pLength,
            s = 10,
            c = 'white',
            **props)
plt.xlim((results['Hela S']['packRatio'].min(), results['Hela S']['packRatio'].max()))
plt.ylim((results['Hela S']['pLength'].min(),   results['Hela S']['pLength'].max()))
plt.xlabel('Packing ratio, bp/nm')
plt.ylabel('Persistance length, nm')
plt.savefig('HelaS_params_means.pdf')
```
