# Tests for Significantly Different Distributions in the STORM/Telomere Project

The major results of the telomere project are probability distributions for measuring a given radius of gyration of a telomere. After knockdown of any particular shelterin protein, we would like to know if the size distribution has changed relative to the control. This means we would like to answer the following question: **do two sets of samples have different probability distributions?**

## Testing the distributions for normality

```python
from IPython.display import Image
Image(filename='fig1c.png')
```

In the box plot above, the mean value is the black dot in the middle of the box, whereas the vertical line in the box middle denotes the median. If the distributions were normal, then these two should lie at the same value for the radius of gyration, but they do not. Therefore, the distributions are probably not normally distributed, but let's double check.

```python
import numpy as np
hl = np.loadtxt('data/Original_Data_L_dataset_RgTrans.txt')
hs = np.loadtxt('data/Original_Data_S_dataset_RgTrans.txt')

# Null hypothesis: the distribution is normal
from scipy.stats.mstats import normaltest
lStats = {}; sStats = {};
_, lStats['normal p-value'] = normaltest(hl)
_, sStats['normal p-value'] = normaltest(hs)

print("""The p-value for the test of whether the Hela L gyration radii are normal is %E.
The p-value for the test of whether the Hela S gyration radii are normal is %E."""
      % (lStats['normal p-value'], sStats['normal p-value']))

print("""Number of Hela L telomeres: %i
Number of Hela S telomeres: %i""" %(hl.size, hs.size))
```

These are incredibly small p-values, so we might at first think that the distributions are likely not normally distributed.

**However**, there is a very big problem with this conclusion. For large sample sizes, most tests of significance will reject the null hypothesis. This means that small deviations from a perfect Gaussian will be preferentially weighted in the normality test when N is large, leading us to extremely small p-values. Here, a normality test should not be thought of as way to determine whether we can use significance tests that require normality; instead, they only tell us whether there's evidence to say that the distribution is not normal.

Due to these complications **a different way to compare two distributions might be required for this project.**

See [this discussion](http://stats.stackexchange.com/questions/2492/is-normality-testing-essentially-useless) on normality testing and large sample sizes and [this discuss](http://stats.stackexchange.com/questions/2516/are-large-data-sets-inappropriate-for-hypothesis-testing) on significance testing with large samples sizes for more information.

## Looking at the effect of TRF2 knockdowns in Hela S

We've seen in our initial analyses that knocking down the Shelterin protein TRF2 seems to shrink the average radius of gyration, especially in Hela S cells. Let's look more closely to see whether this effect is real.

```python
import matplotlib.pyplot as plt
%matplotlib inline

control = np.loadtxt('data/Pooled_pSuper_S_dataset_RgTrans')
treatment = np.loadtxt('data/Pooled_TRF2_S_dataset_RgTrans')

fig, ax = plt.subplots(1,1, dpi = 300)
myBins = np.arange(10, 200, 5) # bin size is 5 nm
ax.hist(control,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Control',
        normed = True)
ax.hist(treatment,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#AAAAAA',
        alpha = 0.5,
        label = 'Treatment',
        normed = True)
ax.grid(True)
ax.legend()
ax.set_title('Hela S')
ax.set_xlabel('Radius of gyration, nm')
ax.set_ylabel('Normalized frequency')

print('Size of control sample: %i; size of treatment sample: %i' % (control.size, treatment.size))
print('Mean of control: %0.1f nm; mean of treatment: %0.1f nm' % (np.mean(control), np.mean(treatment)))
```

These distributions are much less different than the comparison between wild type Hela L and Hela S primarily because they're both Hela S cells. Now, let's look a little deeper into the differences between the distributions.

First, we see that the means are different: 80 nm for the control vs. 74 nm from the treatment. Looking at the distributions, it appears that the control has higher probability than the treatment in the right-sided tail, whereas the left-sided tail is larger for the treatment. Additionally, if we ignore the tails and look at the main body of the distributions, we may conclude that the body of the Treatment distribution is indeed shifted towards smaller radii of gyration.

Are there more quantitative ways to assess their differences?

### The two sample t-test

As I discussed above these distributions fail normality tests, but normality of the distributions is a requirement for performing a valid two sample t-test.

On the other hand, the two sample t-test is robust against non-normal distributions for large sample sizes, so let's try running it anyway on these distributions.

Note: strictly speaking, we will use Welch's t-test because the variances of the underlying distributions cannot be assumed to be equal. This is set by the *equal_var = False* argument.

```python
from scipy import stats
_, pValue = stats.ttest_ind(control, treatment, equal_var = False)
print("The p-value for Welch's two sample t-test is: %E" % pValue)
```

Here we have a p-value of almost *300 millionths*. The test suggests that the mean values really are different, but given the problem with large sample sizes, is this true?

### The Mann-Whitney-Wilcoxon rank-sum test

The [Mann-Whitney-Wilcoxon rank-sum test](http://en.wikipedia.org/wiki/Mann%E2%80%93Whitney_U_test) is a non-parametric test for the null hypothesis that two samples are drawn from the same distribution. The alternative is that measurements in one sample are more likely to be larger than measurements in the other. This non-parametric test does not require normality; indeed, it has very few assumptions, so let's see what it has to say.

```python
_, pValue = stats.ranksums(control, treatment)
print("The p-value for the Wilcoxon rank-sum test is: %E." % pValue)
```

The rank-sum test produces a p-value that is about one millionth, so it would also tell us to reject the null hypothesis and consider the two distributions as different. The question still remains as to whether we can trust this result.

### Bootstrapping

[Bootstrapping techniques](http://en.wikipedia.org/wiki/Bootstrapping_%28statistics%29) are ways of inferring the accuracy of sample estimates, like the mean. They work by treating the measured distribution as the unknown population. Estimates of the accuracy of the population are then derived by sampling *with replacement* from the measured distribution.

Why would we do this? Well, let's think about why the two sample t-test failed. The t statistic for Welch's two sample t-test for samples with different variances is

\begin{equation*} t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} \end{equation*}

$\bar{X}_1$ and $\bar{X}_2$ are the sample means, $s_1$ and $s_2$ are the sample standard deviations, $n_1$ and $n_2$ are the number of measurements in each of the two samples. Now, if the sample size gets very large, this has the effect of making the denominator of $t$ very small, making $t$ itself large. $t$ is what is used to determine the probability in the tails of Student's distribution: large $t$ values have correspondingly low p-values.

Therefore, **the reason the p-value is so small in the t-test is because $n_1$ and $n_2$ are relatively large.**

The t-test essentially assumes our measurements have no uncertainty, so the only error in estimating the mean value is from undersampling. With large sample sizes, the error from undersampling is small, so the error in the mean estimate should be small. But our measurements have large sources of uncertainty associated with them, and these are not taken into account in the t-test.

To better assess the error in the mean, let's make a bootstrap estimate of its error. We'll first write a bootstrap function that samples from our two distributions with replacement.

Note: See [this page](http://nbviewer.ipython.org/gist/aflaxman/6871948) for one idea about implementing bootstrap in Python. Also, the [Wikipedia page](http://en.wikipedia.org/wiki/Bootstrapping_%28statistics%29#Types_of_bootstrap_scheme) suggests that the size of the number of samples in one resample should equal the size of the original dataset.

```python
def bootstrap(data, numSamples = None, statistic = 'mean'):
    """Takes an array of measurements and resamples them with replacement.
    
    Arguments
    ---------
    data : array of float
        The data array
    numSamples : int (optional)
        How many times the distribution will be resampled.
        
    Returns
    -------
    bsStat : float
        The bootstrapped mean value.
    """
    if numSamples is None:
        numSamples = data.size
        
    randInts = np.random.randint(0, high = data.size, size = numSamples)
    resampledData = data[randInts]
    
    if statistic == 'mean':
        bsStat = np.mean(resampledData)
    elif statistic == 'median':
        bsStat = np.median(resampledData)
    try:
        return bsStat
    except:
        print("Error! Was the statistic argument 'mean' or 'median'?")
        raise
```

```python
bsMeanCtrl = bootstrap(control)
bsMeanTrtm = bootstrap(treatment)
print("""The bootstrapped control mean is: %0.1f nm, the sample mean is: %0.1f nm.
The bootstrapped treatment mean is: %0.1f nm, the sample mean is: %0.1f nm."""
      % (bsMeanCtrl, np.mean(control), bsMeanTrtm, np.mean(treatment)))
```

We see after a bootstrap resampling in which we use the size of the data sets to determine the number of resamples, the bootstrapped mean values are relatively close to the sample means. How does this look if we perform a large number of bootstraps, say, 100,000?

```python
numBootstraps = 100000
bsMeans = {'Control' : np.zeros(numBootstraps), 'Treatment' : np.zeros(numBootstraps)}

for ctr in range(numBootstraps):
    bsMeans['Control'][ctr] = bootstrap(control)
    bsMeans['Treatment'][ctr] = bootstrap(treatment)

figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(70, 90, 0.1) # Use bins 0.1 nm wide
ax.hist(bsMeans['Control'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Control',
        normed = True)
ax.hist(bsMeans['Treatment'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#AAAAAA',
        alpha = 0.5,
        label = 'Treatment',
        normed = True)
ax.grid(True)
ax.legend()
ax.set_title('Bootstrapped mean distributions')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')

print("""Bootstrapped control mean: %0.1f
Bootstrapped treatment mean: %0.1f""" % (np.mean(bsMeans['Control']), np.mean(bsMeans['Treatment'])))
```

These distributions never overlap, so any statistical test for the difference between means would be superfluous. The bootstrap technique would allow us to conclude that the means are *significantly different*.

#### Bootstrapping the difference of means

Even if a significance test would be superfluous, it would be good to establish how to do bootstrapping on the difference of two sample means.

A good example of this solution in R is at [this website](http://spark.rstudio.com/ahmed/bootstrap/).

```python
bsDiffs = bsMeans['Control'] - bsMeans['Treatment']

figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(0, 14, 0.1) # Use bins 0.1 nm wide
ax.hist(bsDiffs,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Difference of means',
        normed = True)
ax.grid(True)
ax.legend()
ax.set_ylim((0, 0.4))
ax.set_title('Bootstrapped difference of the means distribution')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')
```

The above is the bootstrapped estimation for the difference in the mean radius of gyration. From this distribution, we can place confidence intervals on the difference of the means by finding the locations in the tails corresponding to certain alpha.

```python
# alpha = 0.05 --> 95% confidence interval; alpha = 0.01 --> 99% confidence interval
alpha = 0.01

lowerBound = np.percentile(bsDiffs, 100 * (alpha / 2))
upperBound = np.percentile(bsDiffs, 100 * (1 - alpha / 2))

# plot the distribution with percentiles marked
yUpperLim = 0.4
figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(0, 14, 0.1) # Use bins 0.1 nm wide
ax.hist(bsDiffs,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Difference of means',
        normed = True)
ax.plot([lowerBound, lowerBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.plot([upperBound, upperBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.grid(True)
ax.legend()
ax.set_ylim((0, yUpperLim))
ax.set_title('Bootstrapped difference of the means distribution')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')

print('The %0.1f percent confidence interval for the difference of means is: (%0.1f, %0.1f)'
      % (100 * (1 - alpha), lowerBound, upperBound))

# Reject the null hypothesis that the distributions have equal means?
if (0 > lowerBound) and (0 < upperBound):
    print('The %0.1f confidence interval includes 0. ACCEPT the null hypothesis.' % (100 * (1 - alpha)))
else:
    print('The %0.1f confidence interval does not include 0. REJECT the null hypothesis.' % (100 * (1- alpha)))
```

### Resampling without replacement

Bootstrapping involves resampling the measured distributions with replacement. What does resampling without replacement tell us?

One thing to be clear about is that resampling without replacement will always lead to larger estimates for the mean uncertainty. This is because we will have, in general, fewer data points in our resample than in the original sample of measurements.

```python
def resampleNoReplace(data, numSamples):
    """Takes an array of measurements and resamples them without replacement.
    
    Arguments
    ---------
    data : array of float
        The data array
    numSamples : int
        How many times the distribution will be resampled. This must be smaller than data.size.
        
    Returns
    -------
    rsMean : float
        The resampled mean value.
    """
    np.random.shuffle(data)
    try:
        rsMean = np.mean(data[:numSamples])
    except:
        print("Error: numSamples must be smaller than the size of the input data array.")
        
    return rsMean
```

```python
numResamples = 100000 # This could be quite slow on some machines, though I've observed it takes a few seconds
sampleSize = 100 # How many samples to take from the distribution
rsMeans = {'Control' : np.zeros(numResamples), 'Treatment' : np.zeros(numResamples)}

for ctr in range(numResamples):
    rsMeans['Control'][ctr] = resampleNoReplace(control, sampleSize)
    rsMeans['Treatment'][ctr] = resampleNoReplace(treatment, sampleSize)
    
figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(60, 100, 0.1) # Use bins 0.1 nm wide
ax.hist(rsMeans['Control'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Control',
        normed = True)
ax.hist(rsMeans['Treatment'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#AAAAAA',
        alpha = 0.5,
        label = 'Treatment',
        normed = True)
ax.grid(True)
ax.legend()
ax.set_title('Resampling w/o replacement mean distributions')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')
```

These are the distributions of the mean if we select 100 samples without replacement from our two measurement distributions. If we decrease the sample size, we should expect a greater width of the the distributions.

```python
numResamples = 100000 # This could be quite slow on some machines, though I've observed it takes a few seconds
sampleSize = 20 # How many samples to take from the distribution
rsMeans = {'Control' : np.zeros(numResamples), 'Treatment' : np.zeros(numResamples)}

for ctr in range(numResamples):
    rsMeans['Control'][ctr] = resampleNoReplace(control, sampleSize)
    rsMeans['Treatment'][ctr] = resampleNoReplace(treatment, sampleSize)
    
figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(60, 100, 0.1) # Use bins 0.1 nm wide
ax.hist(rsMeans['Control'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Control',
        normed = True)
ax.hist(rsMeans['Treatment'],
        bins = myBins,
        histtype = 'stepfilled',
        color = '#AAAAAA',
        alpha = 0.5,
        label = 'Treatment',
        normed = True)
ax.grid(True)
ax.legend()
ax.set_title('Resampling w/o replacement mean distributions')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')
```

At this point we could ask the following question: at what sample size does subsampling without replacement produce mean gyration radii that are not statistically different?

My hypothesis is that the smaller this value for the sample size, the more likely the means are significantly different. When the sample size is equal to the size of the individual distributions, the resampled mean will be the same as the sample means. The resampled means will be equivalent in this case only if the sample means are equivalent.

On the other hand, if the resample size is one, the distributions should be uniformly distributed over the full range of values in the original sample, and there will be significant over lap between the Control and Treatment distributions. In this case, undersampling should be the main source of error in the mean value.

The lower the size of the resamples for a significantly different result, the less the spread in the mean estimates is due to undersampling and the more it reflects the true dispersion of the real, measured sample means.

```python
"""First find the standard error of the means.
These could reasonably be taken as the standard deviation of the resampled distributions.
"""

S = np.sqrt(2 * (np.std(rsMeans['Control'])**2 + np.std(rsMeans['Control'])**2) / 19)

# Now compute a confidence interval for the difference of means
# The t-value for 38 degrees of freedom ((20 - 1) + (20 - 1)) and 95% confidence is 2.024
# Here's a t-calculator http://onlinestatbook.com/2/estimation/difference_means.html

lBound = np.mean(control) - np.mean(treatment) - 2.024 * S
uBound = np.mean(control) - np.mean(treatment) + 2.024 * S

print('The 99 percent confidence interval for the difference of means is %0.3f, %0.3f' % (lBound, uBound))
```

The 99% confidence interval for resampling the distributions without replacement and a sample size of twenty is around $\left[1.4, 12.2 \right]$ (depending on each run). Therefore, it's not likely that undersampling would cause us to accept the null hypothesis with resamples of size 20. We would *likely conclude that the distribution means are different.*

### Wild type and control -- the difference of means

```python
numBootstraps = 100000
bsMeans = {'Wild type' : np.zeros(numBootstraps), 'Control' : np.zeros(numBootstraps)}

for ctr in range(numBootstraps):
    bsMeans['Wild type'][ctr] = bootstrap(hs)
    bsMeans['Control'][ctr] = bootstrap(control)

bsMeanDiffs = bsMeans['Wild type'] - bsMeans['Control']

# alpha = 0.05 --> 95% confidence interval; alpha = 0.01 --> 99% confidence interval
alpha = 0.01

lowerBound = np.percentile(bsMeanDiffs, 100 * (alpha / 2))
upperBound = np.percentile(bsMeanDiffs, 100 * (1 - alpha / 2))

# plot the distribution with percentiles marked
yUpperLim = 0.5
figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(-10, 10, 0.1) # Use bins 0.1 nm wide
ax.hist(bsMeanDiffs,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Difference of means',
        normed = True)
ax.plot([lowerBound, lowerBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.plot([upperBound, upperBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.grid(True)
ax.legend()
ax.set_ylim((0, yUpperLim))
ax.set_title('Bootstrapped difference of the means distribution')
ax.set_xlabel('Mean radius of gyration, nm')
ax.set_ylabel('Normalized frequency')

print('The %0.1f percent confidence interval for the difference of means is: (%0.1f, %0.1f)'
      % (100 * (1 - alpha), lowerBound, upperBound))

# Reject the null hypothesis that the distributions have equal means?
if (0 > lowerBound) and (0 < upperBound):
    print('The %0.1f confidence interval includes 0. ACCEPT the null hypothesis.' % (100 * (1 - alpha)))
else:
    print('The %0.1f confidence interval does not include 0. REJECT the null hypothesis.' % (100 * (1- alpha)))

```

Let's quickly estimate a p-value for this conclusion. To do this, we shift the distribution by its mean so that its mean is centered on zero. Then, we compute the fraction of this curve that is less than the observed difference of the means.

```python
bsMeanDiffsShifted = bsMeanDiffs - np.mean(bsMeanDiffs)
observedMeanDiff = np.mean(hs) - np.mean(control)

pValue = bsMeanDiffsShifted[bsMeanDiffsShifted < observedMeanDiff].size / bsMeanDiffsShifted.size
_, rsPValue = stats.ranksums(hs, control)
print("""The estimated p-value is: %0.6f
Compare this to the ranksum test p-value: %0.6f""" % (pValue, rsPValue))
```

### Wild type and control -- difference of medians

```python
numBootstraps = 100000
bsMedians = {'Wild type' : np.zeros(numBootstraps), 'Control' : np.zeros(numBootstraps)}

for ctr in range(numBootstraps):
    bsMedians['Wild type'][ctr] = bootstrap(hs, statistic = 'median')
    bsMedians['Control'][ctr] = bootstrap(control, statistic = 'median')

bsMedianDiffs = bsMedians['Wild type'] - bsMedians['Control']

# alpha = 0.05 --> 95% confidence interval; alpha = 0.01 --> 99% confidence interval
alpha = 0.01

lowerBound = np.percentile(bsMedianDiffs, 100 * (alpha / 2))
upperBound = np.percentile(bsMedianDiffs, 100 * (1 - alpha / 2))

# plot the distribution with percentiles marked
yUpperLim = 0.5
figure, ax = plt.subplots(1, 1, dpi = 300)
myBins = np.arange(-10, 10, 0.1) # Use bins 0.1 nm wide
ax.hist(bsMedianDiffs,
        bins = myBins,
        histtype = 'stepfilled',
        color = '#333333',
        label = 'Difference of medians',
        normed = True)
ax.plot([lowerBound, lowerBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.plot([upperBound, upperBound], [0, yUpperLim], 'k--', linewidth = 2)
ax.grid(True)
ax.legend()
ax.set_ylim((0, yUpperLim))
ax.set_title('Bootstrapped difference of the medians distribution')
ax.set_xlabel('Median radius of gyration, nm')
ax.set_ylabel('Normalized frequency')

print('The %0.1f percent confidence interval for the difference of medians is: (%0.1f, %0.1f)'
      % (100 * (1 - alpha), lowerBound, upperBound))

# Reject the null hypothesis that the distributions have equal medians?
if (0 > lowerBound) and (0 < upperBound):
    print('The %0.1f confidence interval includes 0. ACCEPT the null hypothesis.' % (100 * (1 - alpha)))
else:
    print('The %0.1f confidence interval does not include 0. REJECT the null hypothesis.' % (100 * (1- alpha)))
```
