# Analyzing day-to-day variation

The purpose of this notebook is to look at the day-to-day variation in the telomere imaging data. This will allow us to better determine if size changes observed in the shelterin knockdown experiments are due to random chance or some change in the structure of the telomeres.

In these plots, we compare each knockdown to the control transfection, *pSuper*. The top plot is always Hela L, and the bottom is always Hela S. We are **plotting the mean values of the distributions** from each experiment. The **errorbars are the standard error of the mean**, which agree well with the bootstrapped estimates performed elsewhere.

## Setup the data structure

```python
import scipy.io as sio
import day2day as d2d
import matplotlib.pyplot as plt
from numpy import arange

# Make plots in this page
%matplotlib inline

# Read the Matlab data into the Python namespace
mlData = sio.loadmat('matlab.mat', variable_names = 'data')

# List of experiment names to read
expName = {'pSuper' : ['Pooled pSuper',
                       '08-09-2014 pSuper',
                       '09-09-2014 pSuper',
                       '21-10-2014 pSuper',
                       '05-11-2014 pSuper',
                       '17-11-2014 pSuper'],
           'TRF1' :   ['Pooled TRF1',
                       '08-09-2014 TRF1',
                       '09-09-2014 TRF1',
                       '21-10-2014 TRF1',
                       '05-11-2014 TRF1',
                       '17-11-2014 TRF1'],
           'TRF2' :   ['Pooled TRF2',
                       '08-09-2014 TRF2',
                       '09-09-2014 TRF2',
                       '21-10-2014 TRF2',
                       '05-11-2014 TRF2',
                       '17-11-2014 TRF2'],
           'TRF1/2' : ['Pooled TRF1 TRF2',
                       '08-09-2014 TRF1 TRF2',
                       '09-09-2014 TRF1 TRF2',
                       '21-10-2014 TRF1 TRF2',
                       '05-11-2014 TRF1 TRF2',
                       '17-11-2014 TRF1 TRF2']}

stats = {'Hela L' : {'Mean' : {}, 'Median' : {}, 'Std. Err.' : {}},
         'Hela S' : {'Mean' : {}, 'Median' : {}, 'Std. Err.' : {}}}

# Remove pSuper from experiment names to make plot xtick labels
ticklabels = []
for label in expName['pSuper']:
    ticklabels.append(label[:-7])
ticklabels = sorted(ticklabels)
```

```python
# Calculate all means, medians, and standard errors of the means
for name in expName.keys():
    stats['Hela L']['Mean'][name],   stats['Hela L']['Std. Err.'][name]   = d2d.findStats(mlData, expName[name], cellType = 'L', statistic = 'mean')
    stats['Hela L']['Median'][name], _                                    = d2d.findStats(mlData, expName[name], cellType = 'L', statistic = 'median')
    
    stats['Hela S']['Mean'][name],   stats['Hela S']['Std. Err.'][name]   = d2d.findStats(mlData, expName[name], cellType = 'S', statistic = 'mean')
    stats['Hela S']['Median'][name], _                                    = d2d.findStats(mlData, expName[name], cellType = 'S', statistic = 'median')
```

```python
# Return the individual distributions for plotting histograms
distrs = {'Hela L' : {}, 'Hela S' : {}}

for name in expName.keys():
    distrs['Hela L'][name], _ = d2d.findStats(mlData, expName[name], cellType = 'L', statistic = 'identity')
    distrs['Hela S'][name], _ = d2d.findStats(mlData, expName[name], cellType = 'S', statistic = 'identity')
```

## TRF1 knockdowns

### Means

```python
y11 =  d2d.sortStats(stats['Hela L']['Mean']['pSuper']);      y21  = d2d.sortStats(stats['Hela S']['Mean']['pSuper'])
y11e = d2d.sortStats(stats['Hela L']['Std. Err.']['pSuper']); y21e = d2d.sortStats(stats['Hela S']['Std. Err.']['pSuper'])
y12 =  d2d.sortStats(stats['Hela L']['Mean']['TRF1']);        y22 =  d2d.sortStats(stats['Hela S']['Mean']['TRF1'])
y12e = d2d.sortStats(stats['Hela L']['Std. Err.']['TRF1']);   y22e = d2d.sortStats(stats['Hela S']['Std. Err.']['TRF1'])
```

```python
options = {'label2' : 'TRF1',
           'xticks' : ticklabels}
plt.show(d2d.plotStats(y11, y11e, y12, y12e, y21, y21e, y22, y22e, **options))
```

### Medians

```python
y11 =  d2d.sortStats(stats['Hela L']['Median']['pSuper']);      y21  = d2d.sortStats(stats['Hela S']['Median']['pSuper'])
y11e = d2d.sortStats(stats['Hela L']['Std. Err.']['pSuper']);   y21e = d2d.sortStats(stats['Hela S']['Std. Err.']['pSuper'])
y12 =  d2d.sortStats(stats['Hela L']['Median']['TRF1']);        y22 =  d2d.sortStats(stats['Hela S']['Median']['TRF1'])
y12e = d2d.sortStats(stats['Hela L']['Std. Err.']['TRF1']);     y22e = d2d.sortStats(stats['Hela S']['Std. Err.']['TRF1'])
```
